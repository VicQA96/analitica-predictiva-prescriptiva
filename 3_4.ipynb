```python
import pandas as pd

brecha_salarial = pd.read_csv(r'C:\Users\Thiare\OneDrive\Escritorio\PROYECTO 3-4\becha_salarial.csv', sep = ',')
ingresos_medios = pd.read_csv(r'C:\Users\Thiare\OneDrive\Escritorio\PROYECTO 3-4\Ingreso_medio.csv', sep = ';')
postulantes = pd.read_csv(r'C:\Users\Thiare\OneDrive\Escritorio\PROYECTO 3-4\Postulantes.csv', sep = ';')
```

```python
# Pivotar `brecha_salarial` para que los años sean filas
brecha_salarial_long = brecha_salarial.melt(
    id_vars=["Region"],  # Columna que no cambia
    var_name="Año",  # Nombre de la columna que contendrá los años
    value_name="Brecha Salarial (%)"  # Nombre de la columna con los valores
)

# Convertir la columna `Año` a entero
brecha_salarial_long["Año"] = brecha_salarial_long["Año"].astype(int)
```

```python
brecha_salarial_long
```

```python
# Verificar tipo de datos
ingresos_medios["Año"] = ingresos_medios["Año"].astype(int)
postulantes["Año"] = postulantes["Año postulación"].astype(int)  # Renombrar para consistencia
```

```python
# Unir brecha salarial con ingresos medios
dataset_parcial = pd.merge(
    brecha_salarial_long,
    ingresos_medios,
    on=["Año", "Region"],  # Coincidencia por Año y Región
    how="inner"  # Mantener solo años y regiones comunes
)
```

```python
# Unir el resultado anterior con el dataset de postulantes
dataset_final = pd.merge(
    dataset_parcial,
    postulantes,
    on="Año",
    how="inner"  # Coincidencia solo para años comunes
)
```

```python
# Inspeccionar datos faltantes
print(dataset_final.isnull().sum())

# Opcional: Llenar datos faltantes con un valor o eliminar filas incompletas
dataset_final.fillna(0, inplace=True)  # Ejemplo: reemplazar nulos con 0
```
